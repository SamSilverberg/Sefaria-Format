```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string

'''COLOR CODES'''##############################################################################################
def pink(st): 
    output = f"\u001b[48;5;219m{st}\u001b[0m"
    return str(output)
def pinkbright(st): 
    output = f"\u001b[48;5;225m{st}\u001b[0m"
    return str(output)
def violet(st): 
    output = f"\u001b[48;5;147m{st}\u001b[0m"
    return str(output)
def violetbright(st): 
    output = f"\u001b[48;5;189m{st}\u001b[0m"
    return str(output)
def blue(st): 
    output = f"\u001b[48;5;75m{st}\u001b[0m"
    return str(output)
def bluebright(st): 
    output = f"\u001b[48;5;87m{st}\u001b[0m"
    return str(output)
def turquoise(st): 
    output = f"\u001b[48;5;80m{st}\u001b[0m"
    return str(output)
def turquoisebright(st): 
    output = f"\u001b[48;5;122m{st}\u001b[0m"
    return str(output)
def green(st): 
    output = f"\u001b[48;5;78m{st}\u001b[0m"
    return str(output)
def greenbright(st): 
    output = f"\u001b[48;5;120m{st}\u001b[0m"
    return str(output)
def coral(st): 
    output = f"\u001b[48;5;216m{st}\u001b[0m"
    return str(output)
def coralbright(st): 
    output = f"\u001b[48;5;223m{st}\u001b[0m"
    return str(output)
def orange(st): 
    output = f"\u001b[48;5;214m{st}\u001b[0m"
    return str(output)
def yellow(st): 
    output = f"\u001b[48;5;220m{st}\u001b[0m"
    return str(output)
def yellowbright(st): 
    output = f"\u001b[48;5;226m{st}\u001b[0m"
    return str(output)
def italic(st):
    output = f"\033[1m{st}\u001b[0m"
    return str(output)


'''MAKE THE DAF'''##############################################################################################
daf = input('Enter the daf number and side you want to access (ex: 2b): ')

if daf[1] == 'a':
    stop = 'Daf '+daf[0]+'b'
else:
    dafnext = (int(daf[0])+1)
    stop = 'Daf '+str(dafnext)+'a'

with open("Berakhot.txt", encoding="utf8") as b:
    b = b.read()
    daffull = re.search(f'{daf}(.*){stop}', b, re.DOTALL)
    daffull = daffull.group(0) #return right clip
    daffull = daffull[0:-7] #cut off second daf and newline



'''CLEAN AND SPLIT INTO SENTENCES'''###############################################################################
d = daffull.split("\n")  #split at newlines, remove them

formatted = ''
for i in d:
    formatted += i.rstrip().lstrip() #string

readable = re.split('([.!?])', formatted) #split string at punctuation
    
k = 0
for i in range(len(readable)): #merge punctuation back
    if k < (len(readable)-1):
        readable[k] = readable[k]+readable[k+1]
    elif k == len(readable):
        break
    k += 2

k = 1
for i in range(len(readable)): #delete punctuation only pieces
    if k < (len(readable)-1):
        del readable[k]
    elif k == len(readable):
        break
    k += 1

final1 = []
for i in readable:
    final1.append(i.lstrip()) 
    
final = []
for i in final1:
    clean = re.sub("\s+"," ", i)
    clean = re.sub("\s+[)]",")", clean)
    clean = re.sub("[(]+\s+","(", clean)
    clean = re.sub("\s+[]]","]", clean)
    clean = re.sub("[[]+\s+","[", clean)
    clean = re.sub("\s+[.]",".", clean)
    clean = re.sub("\s+[,]",",", clean)
    clean = re.sub("\s+[?]","?", clean)
    clean = re.sub("\s+[:]",":", clean)
    final.append(clean+'\n')
    
'''TOPIC EXTRACTION WHOLE'''##########################################################################################               

clean = ' '.join(final)
clean = re.sub("\n"," ", clean)
clean = re.sub("[^-9A-Za-z ]"," ", clean)
clean = re.sub("\s+"," ", clean)
clean = "".join([i.lower() for i in clean if i not in string.punctuation])
words = re.split('\s', clean)
stop_words = set(stopwords.words('english')) 
words_new = [w for w in words if not w in stop_words]
wn = nltk.WordNetLemmatizer()
lemm_words = [wn.lemmatize(word) for word in words_new]
counts = dict() #occurence counts
for word in lemm_words:
    if word in counts:
        counts[word] += 1
    else:
        counts[word] = 1

biggestcount = 0
for key in counts.keys():
    if counts[key] > biggestcount:
        biggestcount = counts[key]
        word = key
if word != 'rabbi' and word != 'rabbis':
    print(word,':',biggestcount)

biggestcount2 = 0
for key in counts.keys():
    if counts[key] > biggestcount2 and counts[key] < biggestcount:
        biggestcount2 = counts[key]
        word2 = key
if word2 != 'rabbi' and word2 != 'rabbis':
    print(word2,':',biggestcount2)

biggestcount3 = 0
for key in counts.keys():
    if counts[key] > biggestcount3 and counts[key] < biggestcount2:
        biggestcount3 = counts[key]
        word3 = key
if word3 != 'rabbi' and word3 != 'rabbis':
    print(word3,':',biggestcount3)

biggestcount4 = 0
for key in counts.keys():
    if counts[key] > biggestcount4 and counts[key] < biggestcount3:
        biggestcount4 = counts[key]
        word4 = key
if word4 != 'rabbi' and word4 != 'rabbis':
    print(word4,':',biggestcount4)



'''TAG DIFFERENT SPEAKERS'''##########################################################################################
speakers = []
colors = [blue, None, turquoise, None, green, None, yellow, None, orange, None, coral, None, pink, None, violet, None, bluebright, None, turquoisebright, None, greenbright, None, yellowbright, None, coralbright, None, pinkbright, None, violetbright]
k = 0
for i in final:
    if re.search(r'([R]+[a|.]+\w+\s+)', i): 
        multiperson = re.search(r'(.*)(\bRabbi\b\s+\w+)(.*)(^[R]+[a|.]+\w+\s+\w+|^[R]+[a|.]+\w*\s\w*|^[R]+[a|.]+\w+:|\bThe\b\s[R]+[a|.]+\w+|\bthe\b\s[R]+[a|.]+\w+|^[R]+[a|.]+\w+\|\s[R]+[a|.]+\w+\s+\w+|.*\bRabbi\b\s\w+)(.*)', i)
        person = re.search(r'(.*)(^\w+\s+[R]+[a|.]+\w+|\bRabbis\b|^[R]+[a|.]+\w*\s\w*|^[R]+[a|.]+\w+:|\bThe\b\s[R]+[a|.]+\w+|\bthe\b\s[R]+[a|.]+\w+|^[R]+[a|.]+\w+\|\s[R]+[a|.]+\w+\s+\w+|[R]+[a|.]+\w+\s+\w+)(.*)', i)
        if multiperson:
            if multiperson.group(2) not in str(speakers) and multiperson.group(4) not in str(speakers):
                speakers.append(multiperson.group(2))
                speakers.append(multiperson.group(4))
        elif person:
            if person.group(2) not in str(speakers):
                speakers.append(person.group(2))
                
                
'''ASSIGN COLORS FOR SPEAKERS'''##########################################################################################      
si = 0
for index,val in enumerate(colors):
    if index%2==1: 
        colors[index] = speakers[si]
        si +=1
    if si == len(speakers):
        break
        
for speakerindex, speakerval in enumerate(speakers):
    for colorindex, colorval in enumerate(colors):
        if speakerval == colorval:
            speakers[speakerindex] = colors[colorindex-1](speakerval)
for i in speakers:
    print(i)
    
print('\n=========================================================================================\n')

    
'''APPLY COLORS TO FULL TEXT'''##########################################################################################
k = 0
for i in final:
    if re.search(r'([R]+[a|.]+\w+\s+)', i): 
        multiperson = re.search(r'(.*)(\bRabbi\b\s+\w+)(.*)(^[R]+[a|.]+\w+\s+\w+|^[R]+[a|.]+\w*\s\w*|^[R]+[a|.]+\w+:|\bThe\b\s[R]+[a|.]+\w+|\bthe\b\s[R]+[a|.]+\w+|^[R]+[a|.]+\w+\|\s[R]+[a|.]+\w+\s+\w+|.*\bRabbi\b\s\w+)(.*)', i)
        person = re.search(r'(.*)(^\w+\s+[R]+[a|.]+\w+|\bRabbis\b|^[R]+[a|.]+\w*\s\w*|^[R]+[a|.]+\w+:|\bThe\b\s[R]+[a|.]+\w+|\bthe\b\s[R]+[a|.]+\w+|^[R]+[a|.]+\w+\|\s[R]+[a|.]+\w+\s+\w+|[R]+[a|.]+\w+\s+\w+)(.*)', i)
        if multiperson:
            prefix = multiperson.group(1)
            name1 = multiperson.group(2)
            between = multiperson.group(3)
            name2 = multiperson.group(4)
            suffix = multiperson.group(5)
            for i, val in enumerate(colors):
                if val == multiperson.group(2):
                    color_tag1 = colors[i-1]
                if val == multiperson.group(4):
                    color_tag2 = colors[i-1]
            newline = prefix+color_tag1(name1)+between+color_tag2(name2)+suffix+'\n'
            final[k] = newline
        elif person:
            for i, val in enumerate(colors):
                if val == person.group(2):
                    color_tag = colors[i-1]
            prefix = person.group(1)
            name = person.group(2)
            suffix = person.group(3)
            newline = prefix+color_tag(name)+suffix+'\n'
            final[k] = newline
    k += 1
for i in final:
    print(i)

```
